```python
import json
```

```python
with open('export.json', 'r') as f:
    data = json.load(f)
```

```python
nested_values = ['video', 'author', 'music', 'stats', 'authorStats','challenges', 'duetInfo', 'textExtra', 'stickerOnItem']
skip_items =['challenges', 'duetInfo', 'textExtra', 'stickerOnItem']
flat_data = {}
for idx, value in enumerate(data):
    flat_data[idx] = {}
    for prop_idx, prop_value, in value.items():
        print(prop_idx)
        if prop_idx in nested_values:
            if prop_idx in skip_items:
                pass
            else:
                for nested_idx, nested_value, in prop_value.items():
                    flat_data[idx][prop_idx+'_'+nested_idx] = nested_value
        else:
            flat_data[idx][prop_idx] = prop_value
```
